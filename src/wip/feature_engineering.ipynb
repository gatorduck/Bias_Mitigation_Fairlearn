# scaling

To do our transformations we will leverage the Scikit-Learns estimators. We start by scaling using min-max scaling, normalization, from 0 to 1.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline

num_pipeline = Pipeline([
    ("scale", MinMaxScaler(feature_range=(0,1))),
])
```

# one hot encoding

Race is a categorical column and needs to be transformed into numeric.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder  

cat_pipeline = Pipeline([
    ("onehot", OneHotEncoder())
])
```

# Putting it all together



```python
from sklearn.compose import ColumnTransformer, make_column_selector

preprocessing = ColumnTransformer([
    ("num", num_pipeline, make_column_selector(dtype_include=['float64'])),
    ("cat", cat_pipeline, make_column_selector(dtype_include=['object'])),
    ("binary", "passthrough", make_column_selector(dtype_include=['int64']))
])
```

```python
X_train_prepped = preprocessing.fit_transform(X_train)

print(X_train_prepped)
```

```python
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin

class CustomCategoricalReplacer(BaseEstimator, TransformerMixin):
    def __init__(self, target_values=None, replacement='Other'):
        self.target_values = target_values
        self.replacement = replacement

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return np.where(X.isin(self.target_values), X, self.replacement)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

# Example data (replace with your actual data)
X = np.array(['white', 'black or african american', 'asian', 'hispanic']).reshape(-1, 1)

# Define the custom categorical replacer
replacer = CustomCategoricalReplacer(target_values=['white', 'black or african american'], replacement='Other')

# Create a ColumnTransformer with the custom replacer and OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ("categorical_replacer", replacer, [0]),  # Replace values in the first column
        ("onehot_encoder", OneHotEncoder(), [0]),  # Apply OneHotEncoder to the same column
        # Add other transformers as needed
    ],
    remainder="passthrough"  # Keep other features unchanged
)

# Fit and transform the data
X_transformed = preprocessor.fit_transform(X)
print(X_transformed)
```
