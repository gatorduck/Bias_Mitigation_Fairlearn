```python
import pickle

pickle_dir = "../data/prepped_data.dat"

with open(pickle_dir, "rb") as f:
    data = pickle.load(f)

X_train_prepped, y_train, X_val, y_val, X_test, y_test, preprocessing = data    

X_train_prepped
```

```python
import pandas as pd

X_val_trans = preprocessing.fit_transform(X_val)


X_val_prepped = pd.DataFrame(
    X_val_trans, columns=preprocessing.get_feature_names_out())

X_val_prepped
```

# Training with Neural Networks

Prep our data for use in our neural network. The framework we will be using is the open source PyTorch.

```python
import torch

# Convert to a tensor
X_train_tensor = torch.tensor(X_train_prepped.values)
y_train_tensor = torch.tensor(y_train.values)
X_val_tensor = torch.tensor(X_val_prepped.values)
y_val_tensor = torch.tensor(y_val.values)

X_train_tensor
```

We will keep this simple and create a neural network with a sigmoid activation function set by our binary outcome. Our loss function is defined by our outcome and will utilize binary cross-entropy.

We define our class first, this sets the stage for layers and types of activation functions we want to use. To create a sequence of layers, we use the Sequential class. You can think of neural networks as nested functions, which each function acting like a layer you can pull back, analagous to an onion.

* activation functions: we use signmoid activation function to return a single value between 0 and 1, and also ReLU 

* loss function: where the learning occurs, give our features with combination of weights what is our error
     - binary cross-entropy

* optimizer: how we move around of loss function, similar to step size, learning rate, etc.
    - gd , sgd, rms prop

```python
import torch
import torch.nn as nn
from torch.optim import RMSprop
from torch.utils.data import DataLoader, TensorDataset

class SimpleNeuralNet(nn.Module):
    def __init__(self):
        super(SimpleNeuralNet, self).__init__()
        self.sequential = torch.nn.Sequential(
            torch.nn.Linear(in_features= 26, out_features=40),  # input layer
            torch.nn.ReLU(),
            torch.nn.Linear(40,40),
            torch.nn.ReLU(),
            torch.nn.Linear(40, 1),
            torch.nn.Sigmoid()
        )

    def forward(self, x):
        x = self.sequential(x)
        return x
    
# Initialize neural network
network = SimpleNeuralNet()

# Define loss function, optimizer
criterion = nn.BCELoss()
optimizer = RMSprop(network.parameters())

# Define data loader
train_data = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_data, batch_size=100, shuffle=True)

# Compile the model using torch 2.0's optimizer
network = torch.compile(network)

# Train neural network
epochs = 3
for epoch in range(epochs):
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = network(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
    print("Epoch:", epoch+1, "\tLoss:", loss.item())

# Evaluate neural network
with torch.no_grad():
    output = network(X_val_tensor)
    test_loss = criterion(output, y_val_tensor)
    test_accuracy = (output.round() == y_val_tensor).float().mean()
    print("Test Loss:", test_loss.item(), "\tTest Accuracy:",
        test_accuracy.item())
```
