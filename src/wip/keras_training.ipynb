lets load lour data from our last step

```python
import pickle

pickle_dir = "../data/prepped_data.dat"

with open(pickle_dir, "rb") as f:
    data = pickle.load(f)

X_train_prepped, y_train, X_val, y_val, X_test, y_test, preprocessing = data    

X_train_prepped
```

```python
X_train_prepped.shape
```

lets transform our validation set using the estimator

```python
import pandas as pd

X_val_trans = preprocessing.fit_transform(X_val)


X_val_prepped = pd.DataFrame(
    X_val_trans, columns=preprocessing.get_feature_names_out())

X_val_prepped
```

# Training with Neural Networks

Prep our data for use in our neural network. The framework we will be using is TensorFlow. To prevent overfitting on such a small dataset we start of with less neurons.

Call compile method which includes loss function and optimizer. This is set by the type of problem we are trying to solve, in this case binary classification. 

https://keras.io/api/losses/
https://keras.io/api/optimizers/

```python
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
import numpy as np

tf.keras.utils.set_random_seed(812) # set seed for weights and other random seeds (e.g. np)

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(25,)), # 1 dimensional array with 25 features
    tf.keras.layers.Dense(40, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

print(model.summary())

model.compile(loss="binary_crossentropy", # loss function, or function for which we will use to minimize our error
              optimizer="RMSprop",  # backpropogation on the loss function performed using RMSprop
              metrics=["accuracy"
                      # ,tf.keras.metrics.AUC()
                      # ,tf.keras.metrics.FalseNegatives()
                      # ,tf.keras.metrics.FalsePositives()
                      # ,tf.keras.metrics.Precision()
                      # ,tf.keras.metrics.Recall()])
])

history  = model.fit(X_train_prepped, y_train, epochs=50, validation_data=(X_val_prepped, y_val))

pd.DataFrame(history .history).plot(
    figsize=(16, 10), xlim=[0, 49], ylim=[0, 1], grid=True, xlabel="Epoch",
    style=["r--", "r--.", "b-", "b-*"])
plt.show()
```

```python
model.save("../model/my_model.keras")

# It can be used to reconstruct the model identically.
#reconstructed_model = tf.keras.models.load_model("../model/my_model.keras")
#model = tf.keras.models.load_model("my_keras_model")
```

Fit our training set to our model object.

make new predictions

```python
X_test_trans = preprocessing.fit_transform(X_test)


X_test_prepped = pd.DataFrame(
    X_test_trans, columns=preprocessing.get_feature_names_out())

y_proba = model.predict(X_test_prepped)
y_proba.round(2)
```

```python
model.evaluate(X_test_prepped, y_test)
```

We will keep this simple and create a neural network with a sigmoid activation function set by our binary outcome. Our loss function is defined by our outcome and will utilize binary cross-entropy.

We define our class first, this sets the stage for layers and types of activation functions we want to use. To create a sequence of layers, we use the Sequential class. You can think of neural networks as nested functions, which each function acting like a layer you can pull back, analagous to an onion.

* activation functions: we use signmoid activation function to return a single value between 0 and 1, and also ReLU 

* loss function: where the learning occurs, give our features with combination of weights what is our error
     - binary cross-entropy

* optimizer: how we move around of loss function, similar to step size, learning rate, etc.
    - gd , sgd, rms prop
