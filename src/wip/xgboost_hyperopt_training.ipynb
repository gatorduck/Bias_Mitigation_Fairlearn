lets load lour data from our last step

```python
import pickle

pickle_dir = "../data/prepped_data.dat"

with open(pickle_dir, "rb") as f:
    data = pickle.load(f)

X_train_prepped, y_train, X_val, y_val, X_test, y_test, preprocessing = data    

X_train_prepped.describe()
```

lets transform our validation set using the estimator

```python
import pandas as pd

X_val_trans = preprocessing.fit_transform(X_val)
X_test_trans = preprocessing.fit_transform(X_test)



X_val_prepped = pd.DataFrame(
    X_val_trans, columns=preprocessing.get_feature_names_out())

X_test_prepped = pd.DataFrame(
    X_test_trans, columns=preprocessing.get_feature_names_out())

X_val_prepped
```

# Training with Xgboost 

```python
import hyperopt import STATUS_OK
import xgboost as xgb
from sklearn.metrics import accuracy_score, recall_score, precision_score, roc_auc_score

def objective(params):


    clf = xgb.XGBClassifier(objective = "binary:logistic"
                            , learning_rate = params['learning_rate']
                            , max_depth = int(params['max_depth'])
                            , tree_method = "hist"
                            , early_stopping_rounds = 5
                            , eval_metric = ["auc","logloss"]
                            , random_state  = 42
    )
                        

    clf.fit(X_train_prepped, y_train, eval_set=[(X_val_prepped, y_val)])

    # clf.best_score

    pred = clf.predict(X_test_prepped)

    acc_metric = accuracy_score(X_test_prepped, pred)

    return ('loss': acc_metric, 'status': STATUS_OK}
```

```python
from hyperopt import hp
import numpy as np

search_space = {
    'learning_rate': hp.loguniform('learning_rate', np.log(1e-3), np.log(1)),
    'max_depth': hp.quniform('max_depth', 2, 10, 2)
}
```

```python
import hyperopt

bayesian_algo = hyperopt.tpe.suggest

random_algo = hyperopt.rand.suggest
```

```python
from hyperopt import fmin, Trials, SparkTrials

trials = SparkTrials(parrallelism=2)

best_param = fmin(fn = objective,
                  space = search_space,
                  algo = bayesian_algo,
                  max_evals = 3,
                  trials = trials)
```

```python
y_pred = clf.predict(X_test_prepped)

print("Accuracy of Model::",accuracy_score(y_test,y_pred))
```
